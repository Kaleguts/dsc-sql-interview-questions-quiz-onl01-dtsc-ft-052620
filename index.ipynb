Primary Keys are the index columns for a given table. A Foreign Key is the Primary Key for a different table in the database than where the Foreign Key resides. You can join two tables together by setting the Primary Key for column A from Table A equal to its Foreign Key in column A of Table B. 




## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
There are 3 different types of relationships that entities can have in a SQL database: one-to-one, one-to-many, and many-to-many. 

A one-to-one relationship means that each row in Table A relates to only one row in Table B and that each row in Table B relates to only one row in Table A.

A one-to-many relationship means that a row in Table A might relate to zero, one, or many rows in Table B and that many rows in Table B may relate to a row in Table A. 

A many-to-many relationship means that many rows in Table A relate to many rows in Table B.

A one-to-one join joins two tables together where the one-to-one relationship exists. A one-to-many join is very helpful in the area of eCommerce. If a customer places an order with multiple items, this join can match the customer information from our customers table with each order they placed from our orderDetails table. 

Using a many-to-many join is usually not very helpful. without other aggregations. If you join a table called restaurants with a table called offices and you have two offices in a city but 513 restaurants you would have 1026 rows afterwards. Sometimes, however, this sort of join is necessary. We may want to join our orders and products tables together. This would constitute a many-to-many join as many orders may have multiple products and many products may be in multiple orders.
